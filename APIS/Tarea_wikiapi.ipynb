# Descripción de la tarea

* Extraer los resúmenes de artículos de wikipedia que contengan la palabra 'astro' (limitar a máximo 20)
* Crear una variable con 0 o 1 que indique si el resumen contiene alguna de las palabras 'física' o 'ciencia'.
* Crear dataframe con columnas: pageid, titulo, resumen, contiene_palabra

```python
import requests
import pandas as pd
```

```python
# Descripcion de la API de wikipedia tipo rest

buscar_en_titulo = 'astro'
endpoint =  'https://es.wikipedia.org/w/rest.php/v1/search/title'

params = {
    'q' : buscar_en_titulo,
    'limit': 20
}

request_wikiapi = requests.get(endpoint, params = params)
request_wikiapi.status_code
```

```python
#Pasamos el formato a un dataframe de los articulos
df = pd.json_normalize(request_wikiapi.json()['pages'])
df
```

```python
#Obtenemos la información en listas: titulo, pagina
lista_titulos = list(df.title)
lista_id = list(df.id)
lista_extractos = []
diccionario_extractos = {}
```

```python
for pgid, titulo in zip(lista_id, lista_titulos):
    url = f'https://es.wikipedia.org/w/api.php?action=query&format=json&pageids={pgid}&prop=extracts&explaintext=True&exintro=True,'
    request_wiki = requests.get(url)
    data = pd.json_normalize(request_wiki.json()['query']['pages'])
    extract = data.iloc[:,3][0]
    lista_extractos.append(extract)
    
diccionario_extractos = {'titulo': lista_titulos,
                        'page_id': lista_id,
                        'extract': lista_extractos}
df_extractos = pd.DataFrame(diccionario_extractos)
df_extractos
```

```python
#Verificar si existe la palabra "ciencia" o "física" en el extracto
verificar_extracto = []

#Recorrer en cada fila del extracto
for index, row in df_extractos.iterrows():
    #Verificamos en cada extracto
    if 'ciencia' in row['extract'] or 'fisica' in row['extract']:
        verificar_extracto.append(1)
    else:
        verificar_extracto.append(0)
        
verificar_extracto
```

```python
#Agregamos columna extra con la verificacion
df_extractos['contiene_palabra'] = verificar_extracto
df_extractos
```
