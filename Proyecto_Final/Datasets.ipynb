# Limpieza de peleadores

```python
#Primero verificamos a todos los peleadores que se tienen en la base de datos y cuantas veces aparecen 
R_unique = list(df_data.R_fighter.unique())
B_unique = list(df_data.B_fighter.unique())
all_names = R_unique+ B_unique
for name in all_names:
    numero =len(df_data[df_data.B_fighter == name]) + len(df_data[df_data.R_fighter == name])
    print(name, numero)
```

```python
len(all_names)
```

```python
posicion_peleadores = ['R', 'B']
bases_peleador = []
#columnas_datos_peleador = ['_fighter', '_Height_cms', '_Reach_cms', '_Weight_lbs']
for name in all_names:
    df_peleador = df_data[(df_data['R_fighter'] == name) | (df_data['B_fighter'] == name)]
    division = df_peleador.weight_class.unique()
    for clase in division:
        diccionario = {}
        diccionario['nombre'] = name
        diccionario['weight_class'] = clase
        for columna in columnas_datos_peleador[2:]:
            datos_recolectados=[]
            datos_sin_vacios = []
            for posicion in posicion_peleadores:
                df_posicion = df_data[(df_data[posicion+'_fighter'] == name) & (df_data['weight_class'] == clase)]
                
                datos_recolectados.append(df_posicion[posicion+columna].unique())
                datos_limpios= list(filter (None, datos_recolectados))
                datos = np.mean(datos_limpios)
            
            diccionario[columna[1:]] = datos
        bases_peleador.append(diccionario)
```

```python
df_datos_peleadores = pd.DataFrame(bases_peleador)
df_datos_peleadores
```

```python
df_datos_peleadores.isna().sum()
```

```python
lista_falta_altura = list(df_datos_peleadores[df_datos_peleadores['Height_cms'].isna()].nombre)
lista_falta_peso = list(df_datos_peleadores[df_datos_peleadores['Weight_lbs'].isna()].nombre)
lista_falta_alcance = list(df_datos_peleadores[df_datos_peleadores['Reach_cms'].isna()].nombre)
```

```python
promedios = df_datos_peleadores.groupby('weight_class')[['Height_cms', 'Weight_lbs']].mean().round(2)
promedios_dict = promedios.to_dict()
promedios_dict
```

```python
for peleador in lista_falta_altura:
    mapping = promedios_dict['Height_cms']
    df_peleador = df_datos_peleadores[df_datos_peleadores.nombre==peleador]
    division = df_peleador['weight_class'].iloc[0]
    promedio_height = mapping.get(division, None)
    df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Height_cms'] = df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Height_cms'].fillna(promedio_height)
    
df_datos_peleadores.isna().sum()
```

```python
for peleador in lista_falta_peso:
    mapping = promedios_dict['Weight_lbs']
    df_peleador = df_datos_peleadores[df_datos_peleadores.nombre==peleador]
    division = df_peleador['weight_class'].iloc[0]
    promedio_weight = mapping.get(division, None)
    df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Weight_lbs'] = df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Weight_lbs'].fillna(promedio_weight)
    
df_datos_peleadores.isna().sum()
```

```python
alcances = df_datos_peleadores.groupby('Height_cms')['Reach_cms'].mean().round(2).fillna(0)
alcances_dict = alcances.to_dict()
alcances_dict

for peleador in lista_falta_alcance:
    df_peleador = df_datos_peleadores[df_datos_peleadores.nombre==peleador]
    division = df_peleador['Height_cms'].iloc[0]
    promedio_reach = alcances_dict.get(division, None)
    df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Reach_cms'] = df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Reach_cms'].fillna(promedio_reach)
    
df_datos_peleadores.isna().sum()
```

```python
posicion_peleadores = ['R', 'B']
stad_peleador = []
for name in all_names:
    df_peleador = df_data[(df_data['R_fighter'] == name) | (df_data['B_fighter'] == name)]
    division = df_peleador.weight_class.unique()
    fechas_peleas = df_peleador.date.unique()
    for clase in division:
        for fecha in fechas_peleas:
            diccionario = {}
            diccionario ['fecha'] = fecha
            diccionario['nombre'] = name
            diccionario['weight_class'] = clase
            for columna in columnas_peleador_historico[3:]:
                for posicion in posicion_peleadores:
                    df_posicion = df_peleador[(df_peleador['weight_class'] == clase) & (df_peleador['date']== fecha)]
                    valor = df_posicion[posicion+columna].unique()
                    if len(valor) == 0:
                        diccionario[columna[1:]] = 'No especificado'
                    else:
                        diccionario[columna[1:]] = valor[0]
            stad_peleador.append(diccionario)
```
