#### Datos de los peleadores

```python
#Proporción de hombres respecto a mujeres en general
total_peleadores_hombres = pd.read_sql(""" SELECT Genre, COUNT(Name) AS number_of_fighters
                                        FROM data_peleadores
                                        WHERE Genre == 'Male'
                                        """, conn)
total_peleadores_mujeres = pd.read_sql(""" SELECT Genre, COUNT(Name) AS number_of_fighters
                                        FROM data_peleadores
                                        WHERE Genre == 'Women'
                                        """, conn)
total_peleadores = pd.concat([total_peleadores_hombres, total_peleadores_mujeres], axis=0)
total_peleadores
```

```python
#Proporción de hombres respecto a mujeres en las categorías donde compaginan.

#Encontrar las categorías que corresponden tanto a mujeres como a hombres
categorias_hombres = pd.read_sql("""SELECT Weight_class, COUNT(Name) AS number_of_fighters
                    FROM data_peleadores
                    WHERE Genre =='Male'
                    GROUP BY Weight_class""", conn)
categorias_mujeres = pd.read_sql("""SELECT Weight_class, COUNT(Name) AS number_of_fighters
                        FROM data_peleadores
                        WHERE Genre == 'Women'
                        GROUP BY Weight_class
                        """, conn)

concatenated_df = pd.merge(categorias_hombres, categorias_mujeres, on='Weight_class')
concatenated_df = concatenated_df.rename(columns={'number_of_fighters_x': 'number_male', 'number_of_fighters_y': 'number_women'})

ax = concatenated_df.plot.bar(cmap = 'plasma')
ax.set_xticklabels(concatenated_df.Weight_class)
ax.set_xlabel("Categories")
ax.set_ylabel("Fighters")
plt.show()
```

```python
# Distribución total de los peleadores entre las categorias
total_categorias = pd.read_sql(""" SELECT Weight_class, COUNT(Weight_class) AS number_of_fighters
                                FROM data_peleadores
                                GROUP BY Weight_class
                                """, conn)

sorted_df = total_categorias.sort_values(by='number_of_fighters',ascending=False)
ax = sorted_df.plot.bar()
ax.set_xticklabels(sorted_df.Weight_class)
ax.set_xlabel("Categories")
ax.set_ylabel("Fighters")
plt.show()
```

#### Datos de las peleas

```python
#Distribucción del total de peleas en el tiempo (por año)
total_peleas_año = pd.read_sql(""" SELECT Year, Count(ID_pelea) AS total_fights
                                FROM data_peleas
                                GROUP BY Year
                                """, conn)
ax= total_peleas_año.total_fights.plot()
ax.set_xticklabels(total_peleas_año.Year)
ax.set_xlabel("Year")
ax.set_ylabel("Fights")
plt.show()
```

```python
#Distribucción del total de peleas por categorías en los ultimos 3 años.
peleas_ultimos_3 = pd.read_sql(""" SELECT Year, Weight_class, Genre, COUNT(Weight_class) AS total_fights
                                    FROM data_peleas
                                    WHERE Year >= 2019
                                    GROUP BY Year, Weight_class, Genre
                                """, conn)

sns.violinplot(data=peleas_ultimos_3, x="Year", y="total_fights",hue='Genre', split='True', palette='plasma')
```

```python
for year in peleas_ultimos_3.Year.unique():
    df = peleas_ultimos_3[peleas_ultimos_3.Year == year]
    df_group = df.groupby('Weight_class').sum()
    ax = df_group.total_fights.plot.bar(cmap = 'viridis')
    ax.set_title(year)
    ax.set_xlabel("Categories")
    ax.set_ylabel("Fighters")
    plt.show()
```

```python
# Distribucción de las localidades donde se realizan las peleas.
lugares_peleas = pd.read_sql("""SELECT Location, COUNT(Location) AS number_fights
                                FROM data_peleas
                                GROUP BY Location
                                ORDER BY number_fights DESC
                                LIMIT 25;
                                """, conn)
mas_peleas = lugares_peleas 
ax = sns.barplot(data=mas_peleas, y="Location", x="number_fights", palette='spring', orient='h')
ax.set_xlabel("Fights")
ax.set_ylabel("Location")
plt.show()
```

```python
#Promedio de knockout por categoría y genero
stad_kd = pd.read_sql(""" SELECT Genre, Weight_class,
                          AVG(Avg_KD) AS avg_kd,
                          AVG(Avg_opp_KD)  AS avg_opp_kd
                          FROM data_estadistica_peleador
                          GROUP BY Weight_class, Genre
                        """, conn)
stad_kd
sns.pairplot(stad_kd, hue="Genre", height=1.5, palette='plasma');
```

```python
stad_laded = pd.read_sql (""" SELECT Genre, Weight_class,
                            AVG(Avg_HEAD_landed) AS HEAD,
                            AVG(Avg_BODY_landed) AS BODY,
                            AVG(Avg_LEG_landed) AS LEG,
                            AVG(Avg_CLINCH_landed) AS CLINCH,
                            AVG(Avg_GROUND_landed) AS GROUND
                            FROM data_estadistica_peleador
                            GROUP BY Weight_class, Genre""", conn)
stad_laded
sns.pairplot(stad_laded, hue="Genre", height=1.5, palette='plasma');
```
