# Exploración de la base de datos
Analizar el tipo de datos que se tiene por columnas, cuantos son valores nulos y si es necesario hacer una conversion de los valores

### Separación de categorías por genero y  comprobación de las categorías respecto al peso

```python
df_data.weight_class.unique()
```

```python
genre = []

categorias = list(df_data.weight_class)
for item in categorias:
    if "Women" in item:
        genre.append('Women')
    else:
        genre.append('Male')
        
df_data['Genre'] = genre
```

```python
df_data.Genre.value_counts()
```

```python
diccionario_categorias={
    'Strawweight': {'min': 113.0, 'max': 116.0},
    'Flyweight': {'min': 123.0, 'max': 126.0},
    'Bantamweight': {'min': 133.0, 'max': 136.0},
    'Featherweight' : {'min': 143.0, 'max': 146.0},
    'Lightweight':{'min': 153.0,'max': 156.0},
    'Welterweight':{'min':170.0, 'max':176.0},
    'Middleweight':{'min': 183.0, 'max': 186.0},
    'Light heavyweight ':{'min':203.0, 'max':206.0},
    'Heavyweight': {'min': 207.0, 'max': 266.0}
}
```

```python
categorias = df_data.loc[:, 'weight_class']
peso_r = df_data.loc[:, 'R_Weight_lbs']
peso_b = df_data.loc[:, 'B_Weight_lbs']
pesos_correcion = []

for categoria, r, b in zip(categorias, peso_r, peso_b):    
    if r == b:
        if  r<=116.0:
            correcion ='Strawweight'
        elif 123.0<=r and r<=126.0:
            correcion = 'Flyweight'
        elif 133.0<=r and r<=136.0:
            correcion = 'Bantamweight'
        elif 143.0<=r and r<=146.0:
            correcion = 'Featherweight'
        elif 153.0<=r and r<=156.0:
            correcion = 'Lightweight'
        elif 170.0<=r and r<=176.0:
            correcion = 'Welterweight'
        elif 183.0<=r and r<=186.0:
            correcion = 'Middleweight'
        elif 203.0<=r and r<=206.0:
            correcion = 'Light heavyweight'
        elif 207.0<=r and r<=266.0:
            correcion = 'Heavyweight'
    elif r-b >= 10:
        correcion ='Peso_Pactado'
    else:
        correcion = 'Error'
    pesos_correcion.append(correcion)

df_data.loc[:,'weight_class'] = pesos_correcion
```

# Limpieza de peleadores

```python
len(sorted(all_names))
```

```python
base_peleador_principal = pd.DataFrame()
base_peleador_principal['Name'] = all_names
base_peleador_principal['ID'] = base_peleador_principal['Name'].map(diccionario_id_peleadores)
```

```python
base_peleador_principal
```

```python
posicion_peleadores = ['R', 'B']
bases_peleador = []
#columnas_datos_peleador = ['_fighter','weight_class', 'Genre', '_Height_cms', '_Reach_cms', '_Weight_lbs']
for name in all_names:
    df_peleador = df_data[(df_data['R_fighter'] == name) | (df_data['B_fighter'] == name)]
    division = df_peleador.weight_class.unique()
    genre = df_peleador.Genre.unique()
    for clase in division:
        diccionario = {}
        diccionario['nombre'] = name
        diccionario['weight_class'] = clase
        diccionario['Genre'] = genre[0]
        for columna in columnas_datos_peleador[3:]:
            datos_recolectados=[]
            datos_sin_vacios = []
            for posicion in posicion_peleadores:
                df_posicion = df_data[(df_data[posicion+'_fighter'] == name) & (df_data['weight_class'] == clase)]
                
                datos_recolectados.append(df_posicion[posicion+columna].unique())
                datos_limpios= list(filter (None, datos_recolectados))
                datos = np.mean(datos_limpios)
            
            diccionario[columna[1:]] = datos
        bases_peleador.append(diccionario)
```

```python
df_datos_peleadores = pd.DataFrame(bases_peleador)
df_datos_peleadores['ID_peleador'] = df_datos_peleadores['nombre'].map(diccionario_id_peleadores)
df_datos_peleadores
```

```python
df_datos_peleadores.isna().sum()
```

```python
lista_falta_altura = list(df_datos_peleadores[df_datos_peleadores['Height_cms'].isna()].nombre)
lista_falta_peso = list(df_datos_peleadores[df_datos_peleadores['Weight_lbs'].isna()].nombre)
lista_falta_alcance = list(df_datos_peleadores[df_datos_peleadores['Reach_cms'].isna()].nombre)
```

```python
promedios = df_datos_peleadores.groupby('weight_class')[['Height_cms', 'Weight_lbs']].mean().round(2)
promedios_dict = promedios.to_dict()
promedios_dict
```

```python
for peleador in lista_falta_altura:
    mapping = promedios_dict['Height_cms']
    df_peleador = df_datos_peleadores[df_datos_peleadores.nombre==peleador]
    division = df_peleador['weight_class'].iloc[0]
    promedio_height = mapping.get(division, None)
    df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Height_cms'] = df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Height_cms'].fillna(promedio_height)
    
df_datos_peleadores.isna().sum()
```

```python
for peleador in lista_falta_peso:
    mapping = promedios_dict['Weight_lbs']
    df_peleador = df_datos_peleadores[df_datos_peleadores.nombre==peleador]
    division = df_peleador['weight_class'].iloc[0]
    promedio_weight = mapping.get(division, None)
    df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Weight_lbs'] = df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Weight_lbs'].fillna(promedio_weight)
    
df_datos_peleadores.isna().sum()
```

```python
alcances = df_datos_peleadores.groupby('Height_cms')['Reach_cms'].mean().round(2).fillna(0)
alcances_dict = alcances.to_dict()
alcances_dict

for peleador in lista_falta_alcance:
    df_peleador = df_datos_peleadores[df_datos_peleadores.nombre==peleador]
    division = df_peleador['Height_cms'].iloc[0]
    promedio_reach = alcances_dict.get(division, None)
    df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Reach_cms'] = df_datos_peleadores.loc[df_datos_peleadores.nombre == peleador, 'Reach_cms'].fillna(promedio_reach)
    
df_datos_peleadores.isna().sum()
```

# Limpieza de datos historicos del peleador

```python
#columnas_peleador_historico = ['date', 'weight_class', 'Genre','_fighter','_total_time_fought(seconds)','_total_rounds_fought','_total_title_bouts','_current_win_streak',
# '_current_lose_streak','_longest_win_streak','_wins','_losses','_draw','_win_by_Decision_Majority','_win_by_Decision_Split',
# '_win_by_Decision_Unanimous','_win_by_KO/TKO','_win_by_Submission','_win_by_TKO_Doctor_Stoppage','_Stance']
```

```python
posicion_peleadores = ['R', 'B']
peleador_historico = []
for name in all_names:
    df_peleador_r = df_data[df_data['R_fighter'] == name]
    df_peleador_b = df_data[df_data['B_fighter'] == name]
    df_posiciones = [df_peleador_r, df_peleador_b]
    df_peleador = df_data[(df_data['R_fighter'] == name) | (df_data['B_fighter'] == name)]
    division = df_peleador.weight_class.unique()
    genre = df_peleador.Genre.unique()
    for posicion, df_posicion in zip (posicion_peleadores, df_posiciones):
        for categoria in division:
            df_temp = df_posicion[df_posicion['weight_class'] == categoria]
            peleas = df_temp['date']
            for pelea in peleas:
                diccionario = {}
                diccionario['nombre'] = name
                diccionario['weight_class'] = categoria
                diccionario['Genre'] = genre[0]
                diccionario['Fight'] = pelea
                año = pelea.split('-')
                diccionario['Year'] = año[0]
                for columna in columnas_peleador_historico[5:]:
                    dato = df_temp[posicion+columna].unique()
                    diccionario[columna[1:]] = dato[0]
                peleador_historico.append(diccionario)
```

```python
df_peleador_historico = pd.DataFrame(peleador_historico)
df_peleador_historico
```

```python
df_peleador_historico.isna().sum()
```

```python
#columnas_stad_peleador= ['_avg_KD', '_avg_opp_KD', '_avg_SIG_STR_pct', '_avg_opp_SIG_STR_pct', '_avg_TD_pct','_avg_opp_TD_pct',
# '_avg_SUB_ATT', '_avg_opp_SUB_ATT', '_avg_REV', '_avg_opp_REV', '_avg_SIG_STR_att', '_avg_SIG_STR_landed','_avg_opp_SIG_STR_att',
# '_avg_opp_SIG_STR_landed','_avg_TOTAL_STR_att','_avg_TOTAL_STR_landed','_avg_opp_TOTAL_STR_att','_avg_opp_TOTAL_STR_landed',
# '_avg_TD_att','_avg_TD_landed','_avg_opp_TD_att','_avg_opp_TD_landed','_avg_HEAD_att','_avg_HEAD_landed','_avg_opp_HEAD_att',
# '_avg_opp_HEAD_landed','_avg_BODY_att','_avg_BODY_landed','_avg_opp_BODY_att','_avg_opp_BODY_landed','_avg_LEG_att',
# '_avg_LEG_landed','_avg_opp_LEG_att','_avg_opp_LEG_landed','_avg_DISTANCE_att','_avg_DISTANCE_landed','_avg_opp_DISTANCE_att',
# '_avg_opp_DISTANCE_landed','_avg_CLINCH_att','_avg_CLINCH_landed','_avg_opp_CLINCH_att','_avg_opp_CLINCH_landed',
# '_avg_GROUND_att','_avg_GROUND_landed','_avg_opp_GROUND_att','_avg_opp_GROUND_landed','_avg_CTRL_time(seconds)',
# '_avg_opp_CTRL_time(seconds)']
```

```python
stad_peleador = []
for name in all_names:
    df_peleador_r = df_data[df_data['R_fighter'] == name]
    df_peleador_b = df_data[df_data['B_fighter'] == name]
    df_posiciones = [df_peleador_r, df_peleador_b]
    df_peleador = df_data[(df_data['R_fighter'] == name) | (df_data['B_fighter'] == name)]
    division = df_peleador.weight_class.unique()
    genre = df_peleador.Genre.unique()
    for posicion, df_posicion in zip (posicion_peleadores, df_posiciones):
        for categoria in division:
            df_temp = df_posicion[df_posicion['weight_class'] == categoria]
            peleas = df_temp['date']
            for pelea in peleas:
                diccionario = {}
                diccionario['nombre'] = name
                diccionario['weight_class'] = categoria
                diccionario['Genre'] = genre[0]
                diccionario['Fight'] = pelea
                año = pelea.split('-')
                diccionario['Year'] = año[0]
                id_pelea= df_temp['ID_peleas'].unique()
                diccionario['ID_pelea'] = id_pelea[0]
                for columna in columnas_stad_peleador:
                    dato = df_temp[posicion+columna].unique()
                    diccionario[columna[1:]] = dato[0]
                stad_peleador.append(diccionario)
```
