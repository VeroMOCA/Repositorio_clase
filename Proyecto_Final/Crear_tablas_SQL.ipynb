```python
import sqlite3 as sql
import csv
import pandas as pd
```

```python
path = './'
connection = sql.connect(path + 'UFC_db.sqlite')
```

# Tabla principal de peleadores

```python
with open('Datos_principal.csv', 'r') as csvfile:
    csv_file_reader = csv.reader(csvfile,delimiter=',')
    #Skip los titulos
    next(csv_file_reader, None)
    
    #Creamos los campos
    Name = ''
    id_peleador = 0
    
    #Creamos la base de datos
    #Creamos la estructura de la tabla
    create_table_main = """
    CREATE TABLE if not Exists data_main(
    ID_peleador INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT NOT NULL
    );
    """
    cursor = connection.cursor()
    cursor.execute(create_table_main)
    
    #Pasamos los datos del csv a la tabla
    for row in csv_file_reader:
        Name = row[0]
            
        #Insertamos en la tabla
        Insertar_datos_main = f""" INSERT INTO data_main(Name)
       VALUES (?)"""
        valores_main = (Name,)
        cursor.execute(Insertar_datos_main, valores_main)
    connection.commit()
```

# Tabla de datos de peleadores

```python
with open('Data_peleadores.csv', 'r') as csvfile:
    csv_file_reader = csv.reader(csvfile,delimiter=',')
    #Skip los titulos
    next(csv_file_reader, None)
    
    #Creamos los campos
    Name = ''
    Weight_class = ''
    Genre = ''
    Height_cms = 0
    Reach_cms = 0
    Weight_lbs = 0
    id_peleador = 0
    
    #Creamos la base de datos
    #Creamos la estructura de la tabla
    create_table_fighters = """
    CREATE TABLE if not Exists data_peleadores(
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT NOT NULL,
    Weight_class TEXT,
    Genre TEXT,
    Height_cms INTEGER,
    Reach_cms INTEGER,
    Weight_lbs INTEGER,
    ID_peleador INTEGER NOT NULL,
    FOREIGN KEY(ID_peleador) REFERENCES data_main (ID_peleador)
    );
    """
    cursor = connection.cursor()
    cursor.execute(create_table_fighters)
    
    #Pasamos los datos del csv a la tabla
    for row in csv_file_reader:
        #Saltamos la primera columna
        for i in range(len(row)):
            #Asiginamos cada valor a su campo
            Name = row[0]
            Weight_class = row[1]
            Genre = row[2]
            Height_cms = row[3]
            Reach_cms = row[4]
            Weight_lbs = row[5]
            id_peleador = row[6]
            
        #Insertamos en la tabla
        Insertar_datos = f""" INSERT INTO data_peleadores (Name, Weight_class, Genre, Height_cms, Reach_cms, Weight_lbs, ID_peleador)
       VALUES (?,?,?,?,?,?,?)"""
        valores = (Name, Weight_class, Genre, Height_cms, Reach_cms, Weight_lbs, id_peleador)
        cursor.execute(Insertar_datos,valores)
    connection.commit()
```
