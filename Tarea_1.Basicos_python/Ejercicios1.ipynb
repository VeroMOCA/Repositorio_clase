Usando la lista de correos generada en la clase obtener:
1. Una lista con los diferentes dominios (sin repetir)
2. Crear un nuevo documento indicando el nombre de usuario y número de vocales que contiene el nombre

```python
import pandas as pd
import numpy as np
```

```python
#Leyendo el archivo
#En la configuración quite los headers
#Y se lo agrege posteriormente de manera manual
doc = pd.read_excel('lista_correos.xlsx',header =None)
doc.columns = ['Correo']
doc
```

```python
#Primero eliminare las filas que no tengan datos 
doc_clean = doc.dropna(subset=['Correo'])
doc_clean
```

```python
dominios = [] #Lista para dominios
nombres = [] #Lista para nombres
for correo in doc_clean.Correo:
    #Un problema que aparecio fue que un dominio se repitio por que al final tenia un espacio en blanco
    #Entonces primero elimine los espacios en blanco al final
    correo = str(correo).rstrip(" ") #Convirtiendo a string
    correo_save= correo.split('@') #Sepando el string por el delimitador '@'
    dominios.append(correo_save[1])#Agregando unicamente el dominio a la lista
    nombres.append(correo_save[0])

    #Creando la lista de valores unicos de dominio
dominios_unicos = np.unique(dominios)
print(dominios_unicos)
```

```python
vocales = ['a', 'e', 'i', 'o', 'u'] #lista de vocales
cx= 0
contar_vocales=[]
for nombre in nombres:
    for i in range(0,len(nombre)):
        if nombre[i] in vocales:
            cx +=1
    print("El nombre {}, tiene {} vocales".format(nombre, cx))
    contar_vocales.append(cx)
    cx =0
    
print(contar_vocales)
```

```python
#Creando diccionario
#Primero guardara el nombre de usuario
#Después guardara el numero de vocales asociadas
usuarios_dic = {'nombre': nombres, 'vocales': contar_vocales}
```

```python
#Ahora convertimos el diccionario en un dataframe
usuarios_df = pd.DataFrame.from_dict(usuarios_dic)
usuarios_df
```

```python
#Ahora convertimos del dataframe a un documento nuevo
usuarios_df.to_csv('Resultados.csv', index=None)
```
